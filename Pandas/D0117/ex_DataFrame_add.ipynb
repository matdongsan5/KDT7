```python
## [1-1] 모듈로딩
import pandas as pd
import numpy as np
```

```python
# [1-2] 데이터준비
dataFile = '../DATA/member.csv'
```

[2] 데이터 로딩 및 확인 <hr>

```python
##[2-1] 데이터 로딩
memDF =pd.read_csv(dataFile)
```

```python
##[2-2] 데이터 확인
memDF.info()
print(memDF.head(),'\n',memDF.tail())
```

```python
memDF.describe(include=[object])
```

```python
memDF.describe(exclude=[object])
```

[3] 행 추가 <hr>

```python
# DF.loc[추가할 행인덱스] = 값
memDF.loc[7] = 100
memDF
```

```python
# DF.loc[추가할 행인덱스] = [리스트, 요소를, 컬럼, 수만큼]
#memDF.loc[8] = [9,'태권브이'] 모자라면 에러.
memDF.loc[8] = [9,'태권브이', 100, '대구']
memDF.loc[7]=0
# 존재하는 행인덱스 지정시에 데이터를 변경함.
memDF.tail(3)
```

```python
memDF.loc[9] = [10,'도라에몽', np.nan, np.nan]

memDF.tail(2)
```

[4] 열 추가 <hr>

```python
## -> DF[컬럼명] = 값
memDF['성별'] = '남'
memDF
```

```python
memDF['혈액형'] = ['A', 'B', 'O', 'AB','A', 'B', 'O', 'AB','A', 'B']
memDF
```

[DataFrame 에서 원소 값 변경]<hr>

```python
## 홍길동의 지역을 대구에서 부산으로 
memDF.loc[0,'지역'] = '부산'
memDF
```

```python
#마징가, 박지민 지역 부산/대구 - 캐나다
memDF.loc[[0,4],'지역'] = ['캐나다','미국']
memDF
```
