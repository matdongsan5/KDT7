[3] 행과 컬럼 데이터 선택/추출 <hr>

```python
## [3-1] 행 선택
## 1개 행 선택 df변수명.loc[행인덱스명]  iloc[행인덱스]
irisDF.loc[0]
irisDF.iloc[0]
```

```python
## 짝수 행 선택 df변수명.loc[행인덱스명]  iloc[행인덱스]
irisDF.iloc[::2]
```

```python
irisDF.loc[::2]
```

```python
# - 짝수 행 인덱스 데이터만 선택 : DF변수명.iloc[ 시작행인덱스:끝행인덱스:간격]
```

```python
## 3-2 행인덱스 변경
## 0~149 --> 0번쨰...149번째
new_index=[]
for idx in range(0,150):
    # irisDF.index[idx] = str(irisDF.index[idx])+'번째'
    new_index.append(f'{idx}번째')
print(new_index)

new_index2 = [f"{idx}번째" for idx in range(len(irisDF))]
print(new_index2)
```

```python
## 인덱스변경 DF.index = 새로운 인덱스 리스트
irisDF.index = new_index2
irisDF.index
```

```python
# 3-3 행선택 df객체변수명.loc[]

print( irisDF.loc['0번째'], irisDF.iloc[0], sep='\n\n')
```

[4] 열 선택/ 추출 <hr>

```python
## [4-1] 1개 열 /컬럼 선택 => DF[컬럼명]
#                           DF.컬럼명
irisDF.columns
```

```python
print(type(irisDF['sepal.length']))
irisDF['sepal.length']
```

```python
## 컬럼명 변경
# new_column = []
# for i in irisDF.columns:
#     new_column.append( i.replace('.','_'))

# irisDF.rename(columns={'sepal.length':'sepla_length'}, inplace=True)

new_column =  list(map(lambda x: x.replace('.', '_'),irisDF.columns))

irisDF.columns=new_column
irisDF.columns
```
